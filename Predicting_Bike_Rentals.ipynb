```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
bike_rentals = pd.read_csv('bike_rental_hour.csv')
bike_rentals.head()
```

```python
plt.hist(bike_rentals['cnt'])
plt.show()
```

```python
bike_rentals.corr()
```

```python
def assign_label(hour):
    if hour >= 6 and hour < 12:
        return 1
    elif hour >= 12 and hour < 18:
        return 2
    elif hour >= 18 and hour < 24:
        return 3
    elif hour >= 0 and hour < 6:
        return 4
```

```python
bike_rentals['time_label'] = bike_rentals['hr'].apply(lambda x: assign_label(x))
bike_rentals['time_label'].value_counts()
```

I will be using Mean Squared Error to evaluate performance since the data is continuous and it doesn't matter so much that we see exactly how close the prediction is rather, what is the best prediction.

```python
from sklearn.linear_model import LinearRegression
train = bike_rentals.sample(frac=.8)
test = bike_rentals.loc[~bike_rentals.index.isin(train.index)]
```

```python
from sklearn.metrics import mean_squared_error

features = ['yr','mnth','hr','temp','hum']
lr = LinearRegression()
lr.fit(train[features],train['cnt'])
predict = lr.predict(test[features])

mse = mean_squared_error(test['cnt'],predict)
rmse = np.sqrt(mse)
print(mse)
print(rmse)
```

The error seems to be pretty high here.  The problem is we dont have many columns that correlate strongly with cnt and none are above .4

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(min_samples_leaf = 10, max_depth = 10)
tree.fit(train[features],train['cnt'])
tree_predictions = tree.predict(test[features])
mse = mean_squared_error(test['cnt'],tree_predictions)
rmse = np.sqrt(mse)
print(mse)
print(rmse)
```

The decision tree seems to do a much better job at reducing error. By limiting the depth to 10 and making the minimum samples per leaf 10 we were able to lower the error even further.  Limiting the depth should also help us to limit overfitting the model. 

```python
train_predictions = tree.predict(train[features])
mse = mean_squared_error(train['cnt'],train_predictions)
rmse = np.sqrt(mse)
print(mse)
print(rmse)
```

As you can see here, the error for the training predictions is not much lower than the test predicitons which means this is doing a good job at not overfitting the tree.

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(n_estimators=25, min_samples_leaf=10, max_depth=10)
forest.fit(train[features],train['cnt'])

forest_prediction = forest.predict(test[features])
mse = mean_squared_error(test['cnt'],forest_prediction)
rmse = np.sqrt(mse)
print(mse)
print(rmse)
```

```python
train_predictions = forest.predict(train[features])
mse = mean_squared_error(train['cnt'], train_predictions)
rmse = np.sqrt(mse)
print(mse)
print(rmse)
```

Here we can see that we further increased the performance of the model by using a random forest regressor.  We got the test and training predictions even closer, reducing overfiting, and also got the overall error down using this method.  There are further improvements we could do such as trying different combinations of features and seeing which perform the best.
