```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')

star_wars.head(10)
```

```python
star_wars
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
star_wars
```

```python
yes_no ={'Yes':True, 'No':False}
for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)
    
star_wars.head()
```

```python
Movie_Mapping = {
                "Star Wars: Episode I  The Phantom Menace": True,
                np.nan: False,
                "Star Wars: Episode II  Attack of the Clones": True,
                "Star Wars: Episode III  Revenge of the Sith": True,
                "Star Wars: Episode IV  A New Hope": True,
                "Star Wars: Episode V The Empire Strikes Back": True,
                "Star Wars: Episode VI Return of the Jedi": True
                }
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(Movie_Mapping)
```

```python
star_wars = star_wars.rename(columns={
    'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1',
    'Unnamed: 4':'seen_2',
    'Unnamed: 5':'seen_3',
    'Unnamed: 6':'seen_4',
    'Unnamed: 7':'seen_5',
    'Unnamed: 8':'seen_6',
})
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
    'unnamed: 10':'rating_2',
    'unnamed: 11':'rating_3',
    'unnamed: 12':'rating_4',
    'unnamed: 13':'rating_5',
    'unnamed: 14':'rating_6',
})
star_wars.head()
```

```python
import matplotlib.pyplot as plt

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

# Rankings

You can see that the favorite movie is episode 5 and that in general people seem to favor the orignal 3 movies vs. the prequel trilogy.  You can see that the average rating in the prequel trilogy slowly goes down as the movies go on but none of them have an average rating in the top half.

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

It makes some sense now why episode 5 is the highest rated, it is also the most watched so if someone has only seen 1 movie it is most likely to be episode 5.  Again the original trilogy appears to be the most watched even if it is not the 3 most watched movies.  The order of the rankings and of the most watched almost match with the only difference being episode 1 is the 3rd most watched but the 4th highest rated and episode 4 is the 4th most watched but 3rd highest rated.
